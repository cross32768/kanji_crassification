```python
import os
import json
```

```python
# 元のjsonファイル（漢字から部首に）を開く
with open('../../data/kanjivg_radical/kanji2radical.json') as f:
    kanji2radical = json.load(f)
```

```python
# 漢字から部首にだと少し扱いにくいので、utf-16(hex)で表現された漢字の下４桁（\u****あるいは+U****みたいな形式になるので）を用いる
utf16_to_radical = {}
for kanji in kanji2radical:
    hex_encode = kanji.encode('utf-16').hex()
    # なぜここで逆転させなければならないのかよくわからない
    hex_encode = (hex_encode[-2:] + hex_encode[-4:-2]).upper()
    utf16_to_radical[hex_encode] = kanji2radical[kanji]
```

```python
# これで書き込み
with open('../../data/kanjivg_radical/utf16_to_radical.json', 'w') as f:
    json.dump(utf16_to_radical, f)
```

```python
# うまくいったか確認
with open('../../data/kanjivg_radical/utf16_to_radical.json') as f:
    utf_to_radical = json.load(f)
```
